```python
import csv

# Original log data
log_data = """
Game ID: 6542

(index):121 Round: 1

game_logic.js:52 Computer Chose: scissors

game_logic.js:98 Player Chose:  scissors

winner_scoreboard.js:34 Round 1 Winner: Draw

(index):121 Round: 2

game_logic.js:52 Computer Chose: lizard

game_logic.js:98 Player Chose:  rock

winner_scoreboard.js:34 Round 2 Winner: Player

(index):121 Round: 3

game_logic.js:52 Computer Chose: scissors

game_logic.js:98 Player Chose:  scissors

winner_scoreboard.js:34 Round 3 Winner: Draw

(index):121 Round: 4

game_logic.js:52 Computer Chose: paper

game_logic.js:98 Player Chose:  spock

winner_scoreboard.js:34 Round 4 Winner: Computer

(index):121 Round: 5

game_logic.js:52 Computer Chose: lizard

game_logic.js:98 Player Chose:  rock

winner_scoreboard.js:34 Round 5 Winner: Player

(index):121 Round: 6

game_logic.js:52 Computer Chose: lizard

game_logic.js:98 Player Chose:  paper

winner_scoreboard.js:34 Round 6 Winner: Computer

(index):121 Round: 7

game_logic.js:52 Computer Chose: scissors

game_logic.js:98 Player Chose:  rock

winner_scoreboard.js:34 Round 7 Winner: Player

(index):121 Round: 8

game_logic.js:52 Computer Chose: spock

game_logic.js:98 Player Chose:  spock

winner_scoreboard.js:34 Round 8 Winner: Draw

(index):121 Round: 9

game_logic.js:52 Computer Chose: lizard

game_logic.js:98 Player Chose:  lizard

winner_scoreboard.js:34 Round 9 Winner: Draw

(index):121 Round: 10

game_logic.js:52 Computer Chose: spock

game_logic.js:98 Player Chose:  rock

winner_scoreboard.js:34 Round 10 Winner: Computer
"""

# Split the log data into individual lines
lines = log_data.strip().split('\n')

# Write the data to a CSV file
with open('output.csv', 'a', newline='') as csvfile:
    writer = csv.writer(csvfile)

    # Write headers
    writer.writerow(["Game ID", "Round", "Computer Choice", "Player Choice", "Round Winner"])

    # Initialize variables to store data for each round
    game_id = ""
    round_number = ""
    computer_choice = ""
    player_choice = ""
    round_winner = ""

    # Process each line of the log data
    for line in lines:
        if "Game ID:" in line:
            game_id = line.split(":")[1].strip()
        elif "Round:" in line:
            round_number = line.split(":")[2].split()[0]  # Extracting only the round number
        elif "Computer Chose:" in line:
            computer_choice = line.split(":")[2].strip()
        elif "Player Chose:" in line:
            player_choice = line.split(":")[2].strip()
        elif "Round" in line and "Winner:" in line:
            round_winner = line.split(":")[2].strip()  # Extracting only the winner
            # Write the data for this round to the CSV file
            writer.writerow([game_id, round_number, computer_choice, player_choice, round_winner])
```
